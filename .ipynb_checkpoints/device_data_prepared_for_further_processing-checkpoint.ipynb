# Converting dataframes, putting right Until time

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
mohd_df_satsp_part1 = pd.read_csv("SATSP_setpoints_jun_oct.csv")
mohd_df_satsp_part2 = pd.read_csv("SATSP_setpoints.csv")
mohd_df_rat_part1 = pd.read_csv("RAT_log.csv")
mohd_df_rat_part2 = pd.read_csv("RAT_log_part2.csv")
mohd_df_rat_part3 = pd.read_csv("RATEMP_log_part2.csv")
mohd_df_ratsp_part_1 = pd.read_csv("RATSP_jun_oct.csv")
mohd_df_ratsp_part_2 = pd.read_csv("RATSP_from_oct.csv")
```

```python
mohd_df_satsp = pd.concat([mohd_df_satsp_part1, mohd_df_satsp_part2], ignore_index=True)
mohd_df_rat = pd.concat([mohd_df_rat_part1, mohd_df_rat_part2, mohd_df_rat_part3], ignore_index=True)
mohd_df_ratsp = pd.concat([mohd_df_ratsp_part_1, mohd_df_ratsp_part_2], ignore_index=True)
mohd_df_ratsp
```

```python
mohd_df_satsp.drop(["property_id", "property_obj", "property_addr"], axis=1, inplace=True)
mohd_df_ratsp.drop(["property_id", "property_obj", "property_addr"], axis=1, inplace=True)
mohd_df_ratsp
```

```python
def make_right_df(csv_df):
    csv_df_dict = csv_df.to_dict('records')
    created_at = []
    time = []
    time_until = []
    prefix = []
    zone = []
    device_name = []
    device_prop_name = []
    parameter = []
    for row in csv_df_dict:
        try:
            datetime_instance = datetime.strptime(row["created_at"].replace(" +0300",''), '%Y-%m-%d %H:%M:%S.%f')
        except ValueError:
            try:
                datetime_instance = datetime.strptime(row["created_at"].replace("+03",''), '%Y-%m-%d %H:%M:%S')
            except ValueError:
                datetime_instance = datetime.strptime(row["created_at"].split(".")[0], '%Y-%m-%d %H:%M:%S')
        created_at.append(datetime.date(datetime_instance))
        time.append(datetime.time(datetime_instance))
        time_until.append("Placeholder")
        d_p_n = row["device_prop_name"].split("'")
        prefix.append(d_p_n[0])
        zone.append(d_p_n[1])
        device_name.append(d_p_n[2])
        device_prop_name.append(d_p_n[3])
        parameter.append(d_p_n[4])
    new_df = pd.DataFrame()
    new_df["created_at"] = created_at
    new_df["Time"] = time
    new_df["Time Until"] = time_until
    new_df["property_value"] = csv_df["property_value"]    
    new_df["prefix"] = prefix
    new_df["Zone"] = zone
    new_df["device name"] = device_name
    new_df["device_prop_name"] = device_prop_name
    new_df["parameter"] = parameter
    return new_df
```

```python
right_satsp_df = make_right_df(mohd_df_satsp).copy(deep=True)
right_rat_df = make_right_df(mohd_df_rat).copy(deep=True)
right_ratsp_df = make_right_df(mohd_df_ratsp).copy(deep=True)
right_ratsp_df
```

```python
print(right_satsp_df.duplicated().any())
print(right_rat_df.duplicated().any())
print(right_ratsp_df.duplicated().any())
right_ratsp_df = right_ratsp_df.drop_duplicates()
right_ratsp_df
```

```python
pivoted_satsp_df = right_satsp_df.pivot(index=["created_at", "Time", "Time Until"], columns=["Zone", "device name", "device_prop_name"], values="property_value").copy(deep=True)
pivoted_satsp_df.reset_index(inplace=True, col_level=2)
pivoted_rat_df = right_rat_df.pivot(index=["created_at", "Time", "Time Until"], columns=["Zone", "device name", "device_prop_name"], values="property_value").copy(deep=True)
pivoted_rat_df.reset_index(inplace=True, col_level=2)
pivoted_ratsp_df = right_ratsp_df.pivot(index=["created_at", "Time", "Time Until"], columns=["Zone", "device name", "device_prop_name"], values="property_value").copy(deep=True)
pivoted_ratsp_df.reset_index(inplace=True, col_level=2)
pivoted_ratsp_df
```

```python
def put_right_Until(df):
    for number_row, actual_row in enumerate(df.to_dict('records')):
        try:
            date_of_the_current_data_point = actual_row[('', '', 'created_at')]
            time_of_the_current_data_point = actual_row[('', '', 'Time')]
            property_with_data = df.iloc[number_row][3:].first_valid_index()
            next_data_point_index = df[property_with_data][number_row + 1:].first_valid_index()
            date_of_the_next_data_point = df.loc[next_data_point_index, ('', '', 'created_at')]
            time_of_the_next_data_point = df.loc[next_data_point_index, ('', '', 'Time')]
        except KeyError:
            date_of_the_next_data_point = actual_row[('', '', 'created_at')]
            time_of_the_next_data_point = actual_row[('', '', 'Time')]
        if date_of_the_next_data_point != date_of_the_current_data_point:
            df.loc[number_row, ('', '', 'created_at')] = date_of_the_next_data_point
        if time_of_the_next_data_point.minute < 30:
            if time_of_the_next_data_point.hour <= 9:
                time_until = f'Until 0{time_of_the_next_data_point.hour}:00'
            else:
                time_until = f'Until {time_of_the_next_data_point.hour}:00'
        else:
            if time_of_the_next_data_point.hour <= 8:
                time_until = f'Until 0{time_of_the_next_data_point.hour + 1}:00'
            else:
                time_until = f'Until {time_of_the_next_data_point.hour + 1}:00'
        
        if time_until == "Until 00:00":
            time_until = "Until 24:00"
        
        df.loc[number_row, ('', '', 'Time Until')] = time_until
    
    return df
        

satsp_df_with_until = put_right_Until(pivoted_satsp_df).copy(deep=True)
rat_df_with_until = put_right_Until(pivoted_rat_df).copy(deep=True)
ratsp_df_with_until = put_right_Until(pivoted_ratsp_df).copy(deep=True)
ratsp_df_with_until
```

```python
def sort_by_time(df):
    hours = []
    df_dict = df.to_dict('records')
    for sort_row in df_dict:
        hours.append(int(sort_row[('', '', 'Time Until')][-5:-3]))
    df[('', '', "Hours")] = hours
    df.sort_values(by=[('', '', 'created_at'),('', '', "Hours")], inplace=True,
               ascending = [True, True], ignore_index=True)
    df.drop([('', '', "Hours")], axis=1, inplace=True)
    return df

sorted_satsp_df = sort_by_time(satsp_df_with_until).copy(deep=True)
sorted_rat_df = sort_by_time(rat_df_with_until).copy(deep=True)
sorted_ratsp_df = sort_by_time(ratsp_df_with_until).copy(deep=True)
sorted_ratsp_df
```

```python
sorted_satsp_df.drop([('', '', 'Time')], axis=1, inplace=True)
sorted_rat_df.drop([('', '', 'Time')], axis=1, inplace=True)
sorted_ratsp_df.drop([('', '', 'Time')], axis=1, inplace=True)
sorted_ratsp_df
```

```python
grouped_satsp_df = sorted_satsp_df.groupby([('', '', 'created_at'), ('', '', 'Time Until')])[list(sorted_satsp_df.columns.values[2:])].first().copy(deep=True)
grouped_rat_df = sorted_rat_df.groupby([('', '', 'created_at'), ('', '', 'Time Until')])[list(sorted_rat_df.columns.values[2:])].first().copy(deep=True)
grouped_ratsp_df = sorted_ratsp_df.groupby([('', '', 'created_at'), ('', '', 'Time Until')])[list(sorted_ratsp_df.columns.values[2:])].first().copy(deep=True)
grouped_ratsp_df
```

```python
satsp_df_index_reset = grouped_satsp_df.reset_index().copy(deep=True)
satsp_df_index_reset.fillna(method='ffill', inplace=True)
satsp_df_filled_nas = satsp_df_index_reset.fillna(method='bfill').copy(deep=True)
rat_df_index_reset = grouped_rat_df.reset_index().copy(deep=True)
rat_df_index_reset.fillna(method='ffill', inplace=True)
rat_df_filled_nas = rat_df_index_reset.fillna(method='bfill').copy(deep=True)
ratsp_df_index_reset = grouped_ratsp_df.reset_index().copy(deep=True)
ratsp_df_index_reset.fillna(method='ffill', inplace=True)
ratsp_df_filled_nas = ratsp_df_index_reset.fillna(method='bfill').copy(deep=True)
ratsp_df_filled_nas
```

```python
def days_in_month(month, year):
    months_with_31_days = [1, 3, 5, 7, 8, 10, 12]
    months_with_30_days = [4, 6, 9, 11]
    if month in months_with_31_days:
        return 31
    elif month in months_with_30_days:
        return 30
    elif month == 2:
        if (year % 400 == 0) and (year % 100 == 0):
            return 29
        elif (year % 4 == 0) and (year % 100 != 0):
            return 29
        else:
            return 28
    else:
        print(f"There was an error in month variable, it's: {month}")


def return_full_months_missing(missing_dates, year):
    number_of_missing_dates = {}
    month = 0
    num_of_days = 0
    for index, date in enumerate(missing_dates):
        if date.month != month and index != 0:
            number_of_missing_dates[month] = num_of_days
            num_of_days = 0
        month = date.month
        num_of_days += 1
    full_months = []
    for missing in number_of_missing_dates.keys():
        full_month = days_in_month(missing, year)
        if number_of_missing_dates[missing] == full_month:
            full_months.append(missing)
    return full_months


def get_all_dates_in_year(start, end):
    delta = end - start
    days = [datetime.date(start + timedelta(days=i)) for i in range(delta.days + 1)]
    return days


def add_full_months(df, months_missing, year):
    for month in months_missing:
        if month < 7:
            present_month = 7 - month + 7
        elif month > 7:
            present_month = 7 - (month - 7)
        else:
            present_month = 8
        filling_part = df.loc[df[('', '', "Months")] == present_month].copy()
        num_of_days_month = list(range(days_in_month(month, year) + 1))
        num_of_days_present_month = list(range(days_in_month(present_month, year) + 1))
        absent_days = [i for i in num_of_days_present_month if i not in num_of_days_month]
        for absent_day in absent_days:
            absent_indices = filling_part.index[filling_part[('', '', "Days")] == absent_day].tolist()
            filling_part = filling_part.drop(absent_indices, axis=0)
        num_of_days_month = [j for j in num_of_days_month if j in list(filling_part[('', '', "Days")].unique())]
        num_of_days_present_month = num_of_days_month.copy()
        num_of_days_present_month.reverse()
        new_dates = []
        new_months = []
        new_days = []
        n = 0
        for num_date, old_date in enumerate(filling_part[[('', '', 'created_at')]].values):
            if num_date == 0:
                working_on_day = old_date[0].day
            if old_date[0].day != working_on_day:
                n += 1
                working_on_day = old_date[0].day
            new_date = old_date[0].replace(month=month, day=num_of_days_present_month[n])
            new_dates.append(new_date)
            new_months.append(month)
            new_days.append(num_of_days_present_month[n])
        filling_part[('', '', 'created_at')] = new_dates
        filling_part[('', '', "Months")] = new_months
        filling_part[('', '', "Days")] = new_days
        df = pd.concat([df, filling_part], ignore_index=True)
    return df


def add_single_dates(df, missing_dates):
    for date in missing_dates:
        new_row_for_date = (df.loc[df[('', '', "Months")] == date.month].iloc[0]).copy()
        new_row_for_date[('', '', 'created_at')] = date
        df.loc[len(df)] = new_row_for_date
    return df


def add_rows_for_whole_year(df, year):
    df_dict = df.to_dict('records')
    months_from_created_at = []
    days_from_created_at = []
    for row_number, row in enumerate(df_dict):
        if row[('', '', 'created_at')].year != year:
            df.loc[row_number, ('', '', 'created_at')] = row[('', '', 'created_at')].replace(year=year)
        months_from_created_at.append(row[('', '', 'created_at')].month)
        days_from_created_at.append(row[('', '', 'created_at')].day)
    df[('', '', "Months")] = months_from_created_at
    df[('', '', "Days")] = days_from_created_at
    df.sort_values(by=[('', '', 'created_at'),('', '', 'Time Until')], inplace=True,
               ascending = [True, True], ignore_index=True)
    start_date = datetime(year, 1, 1)
    end_date = datetime(year, 12, 31)
    dates_range = get_all_dates_in_year(start_date, end_date)
    dates_present = list(df[('', '', 'created_at')].unique())
    missing_dates_range = [i for i in dates_range if i not in dates_present]
    full_months_missing = return_full_months_missing(missing_dates_range, year)
    df = add_full_months(df, full_months_missing, year)
    new_dates_present = list(df[('', '', 'created_at')].unique())
    new_missing_dates_range = [i for i in dates_range if i not in new_dates_present]
    df = add_single_dates(df, new_missing_dates_range)
    df.drop([('', '', "Months"), ('', '', "Days")], axis=1, inplace=True)
    df = sort_by_time(df)
    return df
    

satsp_df_whole_year = add_rows_for_whole_year(satsp_df_filled_nas, 2022).copy(deep=True)
rat_df_whole_year = add_rows_for_whole_year(rat_df_filled_nas, 2022).copy(deep=True)
ratsp_df_whole_year = add_rows_for_whole_year(ratsp_df_filled_nas, 2022).copy(deep=True)
ratsp_df_whole_year
```

```python
def copy_previous_row(df, created_at, time_a, time_p, rows_added):
    needed_row = (df[(df[('', '', 'created_at')] == created_at) & (df[('', '', 'Time Until')] == time_p)]).copy()
    if int(time_a[-5:-3]) - int(time_p[-5:-3]) > 1:
        new_time_p = f"Until 0{int(time_a[-5:-3]) - 1}:00" if (int(time_a[-5:-3]) - 1) <= 9 else f"Until {int(time_a[-5:-3]) - 1}:00"
        needed_row = (df[(df[('', '', 'created_at')] == created_at) & (df[('', '', 'Time Until')] == new_time_p)]).copy()
    needed_row_index = needed_row.index.values[0]
    needed_row[('', '', 'Time Until')] = time_a
    needed_row = needed_row.values[0]
    if int(time_a[-5:-3]) > int(time_p[-5:-3]):
        df.loc[needed_row_index + 0.5] = needed_row
    else:
        if needed_row_index == 0:
            df.loc[-1] = needed_row
            df.index = df.index + 1 
        else:
            df.loc[needed_row_index - 0.5] = needed_row
    df = df.sort_index().reset_index(drop=True)
    rows_added += 1
    return df, rows_added


def add_missing_time_and_rows(df):
    df_dict = df.to_dict('records')
    times_must_have = ["Until 01:00", "Until 02:00", "Until 03:00", "Until 04:00", "Until 05:00", "Until 06:00",
                       "Until 07:00", "Until 08:00", "Until 09:00", "Until 10:00", "Until 11:00", "Until 12:00",
                       "Until 13:00", "Until 14:00", "Until 15:00", "Until 16:00", "Until 17:00", "Until 18:00",
                       "Until 19:00", "Until 20:00", "Until 21:00", "Until 22:00", "Until 23:00", "Until 24:00"]
    rows_added = 0
    for index_for_cur_date, created_at in enumerate(df[('', '', 'created_at')].unique()):
        times_present = df.loc[df[('', '', 'created_at')] == created_at][('', '', 'Time Until')].values.tolist()
        times_absent = [j for j in times_must_have if j not in times_present]
        index_of_present = 0
        for time_a in times_absent:
            if int(time_a[-5:-3]) < int(times_present[index_of_present][-5:-3]):
                df, rows_added = copy_previous_row(df, created_at, time_a, times_present[index_of_present], rows_added)
            else:
                while int(time_a[-5:-3]) > int(times_present[index_of_present][-5:-3]):
                    if len(times_present) == 1:
                        break
                    else:
                        times_present.pop(0)
                df, rows_added = copy_previous_row(df, created_at, time_a, times_present[index_of_present], rows_added)
    return df


satsp_df_with_missing_rows = add_missing_time_and_rows(satsp_df_whole_year).copy(deep=True)
rat_df_with_missing_rows = add_missing_time_and_rows(rat_df_whole_year).copy(deep=True)
ratsp_df_with_missing_rows = add_missing_time_and_rows(ratsp_df_whole_year).copy(deep=True)
ratsp_df_with_missing_rows
```

```python
def add_through_and_until_every_second_row(df):
    df_dict = df.to_dict('records')
    row_index = 1
    row_date = 0
    for index, row in enumerate(df_dict):
        month = f"{row[('', '', 'created_at')].month:02d}"
        day = f"{row[('', '', 'created_at')].day:02d}"
        if index == 0:
            row_date = row[('', '', 'created_at')]
            df.loc[-3] = [f'Through: {month}/{day}'] * len(df.columns.values)
            df.loc[-2] = ['For: AllDays'] * len(df.columns.values)
            df.loc[-1] = [row[('', '', 'Time Until')]] * len(df.columns.values)
            df.index = df.index + 3
            row_index += 3
            continue
        if row[('', '', 'created_at')] != row_date:           
            df.loc[row_index - 0.5] = [f'Through: {month}/{day}'] * len(df.columns.values)
            df = df.sort_index().reset_index(drop=True)
            df.loc[row_index + 1 - 0.5] = ['For: AllDays'] * len(df.columns.values)
            df = df.sort_index().reset_index(drop=True)
            row_date = row[('', '', 'created_at')]
            row_index += 2
        df.loc[row_index - 0.5] = [row[('', '', 'Time Until')]] * len(df.columns.values)
        df = df.sort_index().reset_index(drop=True)
        row_index += 2
    return df


df_satsp_final = add_through_and_until_every_second_row(satsp_df_with_missing_rows)
df_rat_final = add_through_and_until_every_second_row(rat_df_with_missing_rows)
df_ratsp_final = add_through_and_until_every_second_row(ratsp_df_with_missing_rows)
df_ratsp_final
```

```python
df_satsp_final.drop([('', '', 'created_at'), ('', '', 'Time Until')], axis=1, inplace=True)
df_rat_final.drop([('', '', 'created_at'), ('', '', 'Time Until')], axis=1, inplace=True)
df_ratsp_final.drop([('', '', 'created_at'), ('', '', 'Time Until')], axis=1, inplace=True)
df_ratsp_final
```

```python
df_satsp_final.to_csv("SATSP_Final.csv", index=False)
df_rat_final.to_csv("RAT_Final.csv", index=False)
df_ratsp_final.to_csv("RATSP_Final.csv", index=False)
```
